```python
world_covid_trainone.rename(columns={'residential_percent_change_from_baseline':'residential_change'},inplace=True)
world_covid_trainone.rename(columns={'retail_and_recreation_percent_change_from_baseline':'recreation_change'},inplace=True)
world_covid_trainone.rename(columns={'grocery_and_pharmacy_percent_change_from_baseline':'pharmacy_change'},inplace=True)
world_covid_trainone.rename(columns={'new_cases_smoothed_per_million':'cases'},inplace=True)
world_covid_trainone.loc[world_covid_trainone['cases']>1000,'cases']=1000


world_covid_trainone['date'] = pd.to_datetime(world_covid_trainone.date)
col_n=['cases','location','date','residential_change',
       'total_cases','temp','aged_65_older','recreation_change',
       'pharmacy_change','total_vaccinations']
world_covid_traintwo=pd.DataFrame(world_covid_trainone,columns=col_n)
world_covid_traintwo
```

```python
import seaborn as sn
corrMatt = world_covid_traintwo[["cases","location","aged_65_older","residential_change",
       "total_cases","temp","recreation_change",
       "pharmacy_change","total_vaccinations"]].corr()
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
sn.heatmap(corrMatt, mask=mask,
           vmax=.8, square=True,annot=True)
```

From the pure correlation, we can't see any relationship, but if we analyze the graph between variables, we can get some obvious results.
